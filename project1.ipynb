```python
import pandas as pd
import json
import matplotlib.pyplot as plt

## Exercice 1.1

with open("users.json", "r") as f:
    data = json.load(f)
df = pd.DataFrame(data)
df.plot.scatter(x=1, y=4)
```

```python
## Exercise 1.3

import numpy as np
import sklearn 
import sklearn.linear_model
import sklearn.metrics

##We have to drop unused data in order to run model.fit, throws error NaN 
df.dropna(subset=['created','karma','submitted'], inplace = True)

## y = ax + b

xs = df["created"]

actual = df["karma"]

model = sklearn.linear_model.LinearRegression()
xs_reshape = np.array(xs).reshape(-1,1)

model.fit(xs_reshape, actual)
```

```python
predicted = model.predict(xs_reshape)

plt.scatter(xs, actual)
plt.plot(xs, predicted, color = 'black')
```

```python
### import numpy as np
import sklearn 
import sklearn.linear_model
import sklearn.metrics

##We have to drop unused data in order to run model.fit, throws error NaN 
df.dropna(subset=['created','karma','submitted'], inplace = True)

## y = ax + by + z

xs = np.array(df[["created","submitted"]])

actual = df["karma"]

model = sklearn.linear_model.LinearRegression()
xs_reshape = np.array(xs).reshape(9299,2)

model.fit(xs_reshape, actual)
```
